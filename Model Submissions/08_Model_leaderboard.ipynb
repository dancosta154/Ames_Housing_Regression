### Modeling

```python
features = ['overall_qual_liv_area', 'overall_qual_squared', 'quality_and_age', 'overall_qual', 'house_total_sq', 'exter_qual_num', 'gr_liv_area', 'kitchen_qual_num', 'bsmt_qual_num', 'garage_area',
           'garage_cars', 'total_bsmt_sf', 'year_remod_or_built', 'year_remod_or_built_sq', 'garage_finish']

X = df[features]
y = df['saleprice']
```

```python
# Train, Test, Split

X_train, X_validate, y_train, y_validate = train_test_split(X, y, random_state=45)
```

```python
# Fit log to account for skewed distribution in saleprice

y_train_log = y_train.map(np.log)
y_validate_log = y_validate.map(np.log)
```

```python
# Instantiate the model

lr = LinearRegression()
```

```python
# Fit the model

lr.fit(X_train, y_train_log)
```

```python
# Train Score

lr.score(X_train, y_train_log)
```

```python
# Test Score

lr.score(X_validate, y_validate_log)
```

```python
# Cross-Val-score

cross_val_score(lr, X_train, y_train_log).mean()
```

### Predictions

```python
preds = lr.predict(X_validate)
preds;
```

```python
# Residuals
residuals = y_validate_log - preds
```

```python
# Plot residuals to evaluate Epsilon
# This exhibits a heteroscedastic distribution, as the points are not quite uniform when scanning left to right
# meaning there is more opportunity to reduce error

plt.axhline(y=0, color='r')
sns.scatterplot(x = preds, y = residuals)
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residuals')
plt.title('Residual Plot showing Reducable Error');
```
