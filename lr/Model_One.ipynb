First run through to submit a model to Kaggle

```python
# import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# read in csv
df = pd.read_csv('../../datasets/train.csv')
df.head()
```

```python
pd.options.display.max_columns = 999
```

```python
df._get_numeric_data().isnull().sum()
```

```python
# Set features equal to Features that I want to model
# Set X = features
# Set y = what we'd like to predict

features = ['Overall Cond', '1st Flr SF']

X = df[features]
y = df['SalePrice']
```

```python
# Instantiate the model

lr = LinearRegression()
```

#### If this score sucks, do not proceed

```python
# Cross-Val-score

cross_val_score(lr, X, y).mean()
```

```python
# Fit the model

lr.fit(X, y)
```

```python
# Train, Test, Split

X_train, X_validate, y_train, y_validate = train_test_split(X, y, random_state=32)
```

```python
# Ensure 75%/25% split

X_train.shape
```

```python
# Ensure 75%/25% split

X_validate.shape
```

```python
# Train Score

lr.score(X_train, y_train)
```

```python
# Test Score

lr.score(X_validate, y_validate)
```

```python
# Cross-Val-score

cross_val_score(lr, X_train, y_train).mean()
```

```python
# R2 score for the Baseline model is 0
```

```python
# Baseline Model - in dollars
baseline = y.mean()
baseline
```

```python
preds = lr.predict(X)
preds
```

```python
# Residuals - Actual minus predictions
residuals = y - preds
```

```python
# Plot residuals to evaluate Epsilon (irreducable error)

plt.axhline(y=0, color='r')
sns.scatterplot(preds, residuals)
plt.ylabel('Predicted Sale Price')
plt.xlabel('Residuals');
```
